```
import pandas as pd
import numpy as np
```

```
data = pd.DataFrame({'X':[25,50,75,100,125], 'Y':[32,52,79,99,119], 'Z':[39,55,82,103,129]})
data
```

<B>PRZYGOTOWANIE TEKSTU</B>

```
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')


x = data["X"]
y = data["Y"]

e =[40,58,85,109,132]
c,d = np.polyfit(x,e, deg=1) 

a, b = np.polyfit(x, y, deg=1)

g = [0,25,50,75,100,125]

y_est= [a*i+b for i in g]
y_est2= [c*i+d for i in g]

y_err = 13
y_err2 = 8
g = pd.Series(g).values
y_est=pd.Series(y_est).values
y_est2=pd.Series(y_est2).values



fig, ax = plt.subplots(figsize=(10, 6))
#1 seria danych
ax.plot(g, y_est, '-')
ax.fill_between(g, y_est - y_err, y_est + y_err, alpha=0.1)
ax.plot(x, y, 'o', color='tab:blue')
ax.set_xlim(0,150)
#2 seria danych
ax.plot(g, y_est2, '--', color='orange')
ax.plot(x, e, 'v', color='tab:orange')
ax.fill_between(g, y_est2 - y_err2, y_est2 + y_err2, alpha=0.1)
ax.set_xlabel('Normal stress [kPa]')
ax.set_ylabel('Shear strength [kPa]')
```
